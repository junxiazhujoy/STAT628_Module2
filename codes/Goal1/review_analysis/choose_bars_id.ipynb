```python
bars=[x for x in dat if ('Bar' in str(x['categories'])) & ('Barbers' not in str(x['categories'])) & ('Sushi Bars' not in str(x['categories'])) & ('Juice Bars' not in str(x['categories'])) & ('Nightlife' in str(x['categories'])) & (x.get('Music', 0)==0) & (x.get('Ambience', 0)==0)]
#bars[0:20]
len(bars)
```

```python
id = {}
for i in range(len(bars)):
    id[i] = bars[i]['business_id']

    
bars_id=pd.Series(list(id.values()))
unique_id = bars_id.unique()
str_bars_id = ''
for ind in unique_id:
    str_bars_id = str_bars_id  + str(ind) + ','
```

```python
f = open('uniq_id_bars.txt','w')
f.write(str_bars_id)
f.close()
```
